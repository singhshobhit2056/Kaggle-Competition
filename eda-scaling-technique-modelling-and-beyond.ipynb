```python
x = newdata.drop('Exited', axis = 'columns')
y = newdata['Exited'] 
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state= 42)
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(x_train, y_train)
```

```python
print('train score model1:',model1.score(x_train, y_train))
print('test score model1:',model1.score(x_test, y_test))
from sklearn.metrics import confusion_matrix, classification_report
cm1 = confusion_matrix(y_train, model1.predict(x_train))
cm2 = confusion_matrix(y_test, model1.predict(x_test))
print('-----------------------------------------------------------')
print('Confusion Matrix 1')
print(cm1)
print('-----------------------------------------------------------')
print('Confusion Matrix 2')
print(cm2)
```
